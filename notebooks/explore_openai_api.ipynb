# Initialize our helper classes

```python
analyst = Analyst()
analyst.load_data()
analyst.build_respondents_list()

plotter = Plotter()

llm = LLM()
```

# Analyze the main skills question

```python
census_question = "In your opinion, what are the top three skills most in demand in the battery industry?"

data_dict = analyst.summarize_census_skills_demand()
string_list = data_dict['skills_in_demand']

model_list = ['gpt-4o', 'o1-preview']

for model in model_list:

    llm_dict, res_dict = llm.analyze(string_list, model=model)

    plotter.make_bar_plot_from_dict(res_dict,
    title=census_question,
    annotation=f'Model: {model}, {time.strftime("%Y/%m/%d %H:%M:%S")}',
    num_elements=50,
    saveas=f'census_skills_{model}_{time.strftime('%Y%m%d_%H%M%S')}.png',
    sorted=True
)

```
